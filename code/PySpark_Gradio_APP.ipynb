```python
# 1. Merge the dataframes
df_merged = pd.merge(df_orders, df_customer, on='Customer ID', how='left')
df_merged = pd.merge(df_merged, df_products, on='Product ID', how='left')

# Handle potential missing values (optional, based on inspection)
# For this task, we assume merges are successful and proceed without explicit missing value handling here.
# If needed, one could use:
# df_merged.dropna(subset=['Customer Name', 'Product Name'], inplace=True)
# Or fill specific columns:
# df_merged['Customer Name'].fillna('Unknown Customer', inplace=True)

# 2. Define the get_customer_orders function
def get_customer_orders(customer_identifier):
    """
    Retrieves the order history for a given customer identifier.

    Args:
        customer_identifier: The Customer ID or Customer Name.

    Returns:
        A DataFrame containing the order history for the customer.
    """
    if customer_identifier in df_merged['Customer ID'].unique():
        customer_orders = df_merged[df_merged['Customer ID'] == customer_identifier]
    elif customer_identifier in df_merged['Customer Name'].unique():
        customer_orders = df_merged[df_merged['Customer Name'] == customer_identifier]
    else:
        return pd.DataFrame({"Message": ["Customer not found."]})

    # Select relevant columns for display
    display_columns = ['Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Product Name',
                       'Category', 'Sub-Category', 'Quantity', 'Price', 'Discount', 'Profit']
    return customer_orders[display_columns]


# 3. Define the get_product_recommendations function
def get_product_recommendations(order_history_df):
    """
    Generates product recommendations based on customer order history.

    Args:
        order_history_df: DataFrame containing the customer's order history.

    Returns:
        A list of recommended product names or a message if no recommendations.
    """
    if order_history_df.empty or "Message" in order_history_df.columns:
        return ["No order history to generate recommendations."]

    # Simple recommendation logic: recommend products from the most frequent category
    category_counts = order_history_df['Category'].value_counts()
    if category_counts.empty:
        return ["Could not determine popular categories for recommendations."]

    most_frequent_category = category_counts.index[0]

    # Find products in the same category that the customer hasn't purchased
    purchased_product_ids = order_history_df['Product ID'].unique()
    recommended_products_df = df_products[
        (df_products['Category'] == most_frequent_category) &
        (~df_products['Product ID'].isin(purchased_product_ids))
    ]

    # Get top N recommended product names (e.g., top 5)
    recommended_product_names = recommended_products_df['Product Name'].head(5).tolist()

    if not recommended_product_names:
        return [f"No new recommendations found in the '{most_frequent_category}' category."]

    return recommended_product_names

# 4. Create a Gradio interface using gr.Blocks()
# 5. Add a gr.Textbox for the user to input the customer identifier.
# 6. Add a gr.Button to trigger the process.
# 7. Add a gr.Dataframe to display the customer's order history.
# 8. Add a gr.Textbox to display the product recommendations.
# 9. Connect components and functions.
# 10. Launch the interface (will be in the next step or within the %%blocks cell).
```

```python
%%blocks
import gradio as gr
import os

# BigQuery client and table IDs will be initialized within the function triggered by the button


with gr.Blocks() as demo:
    gr.Markdown("# Customer Order History (from BigQuery)")
    customer_input = gr.Textbox(label="Enter Customer ID or Customer Name")
    process_button = gr.Button("Get Order History from BigQuery")
    order_history_output = gr.Dataframe(label="Order History")

    def process_customer_data_from_bq(identifier):
        import pandas as pd
        from google.oauth2 import service_account
        from google.cloud import bigquery
        # BigQuery authentication and client initialization inside the function
        CREDENTIALS_PATH = 'cred.json'

        try:
            credentials = service_account.Credentials.from_service_account_file(CREDENTIALS_PATH)
            client = bigquery.Client(credentials=credentials)

            # Replace with your desired Google Cloud Project ID and Dataset ID
            project_id = credentials.project_id  # Use the project ID from your credentials
            dataset_id = 'customer_data' # Replace with your desired dataset ID
            merged_table_id = f"{project_id}.{dataset_id}.customer_orders_products"

        except Exception as e:
            return pd.DataFrame({"Error": [f"Error initializing BigQuery client: {e}"]})

        if client is None:
            return pd.DataFrame({"Error": ["BigQuery client could not be initialized."]})

        # Construct the BigQuery SQL query
        # We will filter by either 'Customer ID' or 'Customer Name'
        query = f"""
        SELECT
            `Order ID`,
            `Order Date`,
            `Ship Date`,
            `Ship Mode`,
            `Product Name`,
            Category,
            `Sub-Category`,
            Quantity,
            Price,
            Discount,
            Profit
        FROM
            `{merged_table_id}`
        WHERE
            `Customer ID` = @identifier
            OR `Customer Name` = @identifier
        """

        # Configure the query job with parameters to prevent SQL injection
        query_job_config = bigquery.QueryJobConfig(
            query_parameters=[
                bigquery.ScalarQueryParameter("identifier", "STRING", identifier),
            ]
        )

        try:
            # Run the query
            query_job = client.query(query, job_config=query_job_config)

            # Fetch the results into a Pandas DataFrame
            customer_orders_df = query_job.to_dataframe()

            if customer_orders_df.empty:
                 return pd.DataFrame({"Message": [f"No orders found for '{identifier}'."]})

            return customer_orders_df

        except Exception as e:
            return pd.DataFrame({"Error": [f"Error querying BigQuery: {e}"]})


    process_button.click(
        fn=process_customer_data_from_bq,
        inputs=customer_input,
        outputs=order_history_output
    )

demo.launch(debug=True)
```

```python
from google.cloud import bigquery

# Define the destination table ID for the merged data
merged_table_id = f"{project_id}.{dataset_id}.customer_orders_products"

# Define the SQL query to join the tables
# We are joining customer, orders, and products tables on their respective IDs.
# Adjust column names and joins based on your actual table schemas in BigQuery if they differ from the DataFrames.
query = f"""
SELECT
    c.`Customer ID`,
    c.`Customer Name`,
    c.email,
    c.phone,
    c.address,
    c.Segment,
    c.Country,
    c.City,
    c.State as Customer_State, -- Alias to distinguish from Product State
    c.`Postal Code`, -- Corrected column name to match the DataFrame/BigQuery table
    c.Region,
    o.`Order ID`,
    o.`Order Date`,
    o.`Ship Date`,
    o.`Ship Mode`,
    o.`Product ID`,
    o.Quantity,
    o.Price,
    o.Discount,
    o.Profit,
    p.Category,
    p.`Sub-Category`,
    p.`Product Name`,
    p.State as Product_State, -- Alias to distinguish from Customer State
    p.`Price per product` -- Corrected column name to match the DataFrame/BigQuery table
FROM
    `{customer_table_id}` AS c
JOIN
    `{orders_table_id}` AS o ON c.`Customer ID` = o.`Customer ID`
JOIN
    `{products_table_id}` AS p ON o.`Product ID` = p.`Product ID`
"""

# Configure the job to create a new table from the query result
job_config = bigquery.QueryJobConfig(destination=merged_table_id)

# Start the query job
print(f"Creating merged table '{merged_table_id}' from query...")
query_job = client.query(query, job_config=job_config)

# Wait for the job to complete
query_job.result()

print(f"Merged table '{merged_table_id}' created and populated.")

# Optional: Verify the created table by querying the first few rows
print(f"\nQuerying merged table '{merged_table_id}' (first 5 rows):")
query = f"SELECT * FROM `{merged_table_id}` LIMIT 5"
query_job = client.query(query)
merged_results = query_job.result()

for row in merged_results:
    print(row)
```
